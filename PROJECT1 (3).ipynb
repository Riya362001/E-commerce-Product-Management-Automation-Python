```python

def load_data(main_folder_path):
  with open('mainfolder/sales_data.csv', 'r') as file:
    csv_reader = csv.DictReader(file)
    sales_data_list = [row for row in csv_reader]
    sales_data={}
    for i in sales_data_list:
      sales_data[i["Product_SKU"]]=[j for j in i.values()][1:]

  product_details_path = 'mainfolder/product_details'


  product_details_list = []
  product_details={}
  for filename in os.listdir(product_details_path):
    if filename.endswith('.json'):
        with open(os.path.join(product_details_path, filename), 'r') as f:
            json_data = json.load(f)
            product_details_list.append(json_data)
  product_details = {index: dictionary for index, dictionary in enumerate(product_details_list )}
  for i, key in enumerate(sales_data):
    product_details[key] = product_details.pop(i)





  directory = 'mainfolder/product_descriptions'

  text_data_list = []
  product_descriptions={}
  for filename in os.listdir(directory):
    if filename.endswith('.txt'):
        with open(os.path.join(directory, filename), 'r') as f:
            text_data = f.read()
            text_data_list.append(text_data)

  product_descriptions = {index: dictionary for index, dictionary in enumerate(text_data_list )}
  for i, key in enumerate(sales_data):
    product_descriptions[key] = product_descriptions.pop(i)

  tuple_of_dicts = (sales_data, product_details, product_descriptions)


  return tuple_of_dicts
```

```python
def update(product_details, sales_data, product_descriptions):
    sku = input("Enter the product SKU to update: ")

    if sku not in product_details:
        print("SKU not found.")
        return product_details, sales_data, product_descriptions

    while True:
        print("\nSelect the data you want to update:")
        print("1. Sales data")
        print("2. Product details")
        print("3. Product description")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":  # Update Sales Data
            new_sales = input("Enter updated quantities for all days (space-separated): ")
            new_sales = new_sales.split()
            if len(new_sales) != 14 or not all(s.isdigit() for s in new_sales):
                print("Invalid input. Please enter exactly 14 integers.")
                continue
            sales_data[sku] = [int(s) for s in new_sales]
            print("The sales data is updated successfully!")

        elif choice == "2":  # Update Product Details
            product_details[sku] = {
                'name': input("Enter the updated product name: "),
                'brand': input("Enter the updated brand: "),
                'model': input("Enter the updated model: "),
                'specifications': input("Enter the updated specifications: "),
                'price': input("Enter the updated price: "),
                'availability': input("Enter the updated availability: ")
            }
            print("The product details are updated successfully!")

        elif choice == "3":  # Update Product Description
            product_descriptions[sku] = input("Enter the updated product description: ")
            print("The product description is updated successfully!")

        elif choice == "4":  # Exit
            print("Exiting...")
            break

        else:
            print("Invalid choice. Please enter a number from 1 to 4.")

    return product_details, sales_data, product_descriptions

```

```python
update(product_details, sales_data, product_descriptions)
```

```python
def delete(sales_data, product_details, product_descriptions, main_folder):
    sku = input("Enter the SKU of the product to be deleted: ")

    if sku not in sales_data:
        print(f"SKU '{sku}' not found.")
        return sales_data, product_details, product_descriptions
    del sales_data[sku]
    del product_details[sku]
    del product_descriptions[sku]

    json_path = os.path.join(main_folder, "product_details", f"{sku}.json")
    if os.path.exists(json_path):
        os.remove(json_path)
        print(f"JSON file for SKU '{sku}' deleted successfully.")
    else:
        print(f"JSON file for SKU '{sku}' not found.")

    # Delete text file
    text_path = os.path.join(main_folder, "product_descriptions", f"{sku}.txt")
    if os.path.exists(text_path):
        os.remove(text_path)
        print(f"Text file for SKU '{sku}' deleted successfully.")
    else:
        print(f"Text file for SKU '{sku}' not found.")

    print(f"Entries for SKU '{sku}' have been deleted successfully.")

    return sales_data, product_details, product_descriptions

```

```python
main_folder = "mainfolder"
sales_data, product_details, product_descriptions = delete(sales_data, product_details, product_descriptions, main_folder)
```

```python

def dump_data(sales_data, product_details, product_descriptions, main_folder):
    product_details_folder = os.path.join(main_folder, "product_details")
    product_descriptions_folder = os.path.join(main_folder, "product_descriptions")

    if not os.path.exists(product_details_folder):
        os.makedirs(product_details_folder)
    if not os.path.exists(product_descriptions_folder):
        os.makedirs(product_descriptions_folder)
    sales_csv_path = os.path.join(main_folder, "sales_data.csv")
    with open(sales_csv_path, "w", newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["SKU", "Day 1", "Day 2", "Day 3", "Day 4", "Day 5", "Day 6", "Day 7", "Day 8", "Day 9", "Day 10", "Day 11", "Day 12", "Day 13", "Day 14"])
        for sku, sales in sales_data.items():
            writer.writerow([sku] + sales)

    for sku, details in product_details.items():
        json_path = os.path.join(product_details_folder, f"{sku}.json")
        with open(json_path, "w") as jsonfile:
            json.dump(details, jsonfile, indent=4)

    for sku, description in product_descriptions.items():
        txt_path = os.path.join(product_descriptions_folder, f"{sku}.txt")
        with open(txt_path, "w") as txtfile:
            txtfile.write(description)
```

```python
dump_data(sales_data, product_details, product_descriptions, main_folder)
```

```python
def user_menu(main_folder):
    while True:
        # Load existing product data
        sales_data, product_details, product_descriptions = load_data(main_folder)

        # Display menu options
        print("\nProduct Data Management Menu:")
        print("1. Create new product data")
        print("2. Read existing data")
        print("3. Update current data")
        print("4. Delete data")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == "1":
            product_details, sales_data, product_descriptions = create(product_details, sales_data, product_descriptions)
            dump_data(sales_data, product_details, product_descriptions, main_folder)

        elif choice == "2":
            read(product_details, sales_data, product_descriptions)

        elif choice == "3":
            product_details, sales_data, product_descriptions = update(product_details, sales_data, product_descriptions)
            dump_data(sales_data, product_details, product_descriptions, main_folder)

        elif choice == "4":
            product_details, sales_data, product_descriptions = delete(product_details, sales_data, product_descriptions, main_folder)
            dump_data(sales_data, product_details, product_descriptions, main_folder)

        elif choice == "5":
            dump_data(sales_data, product_details, product_descriptions, main_folder)
            print("Exiting...")
            break

        else:
            print("Invalid choice. Please enter a number from 1 to 5.")
```

```python
user_menu("mainfolder")
```
